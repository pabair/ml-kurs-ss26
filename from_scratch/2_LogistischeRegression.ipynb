```python
import numpy as np

class LogisticRegression:
    def __init__(self, learning_rate=0.01, iterations=1000, tolerance=1e-6):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.tolerance = tolerance
        self.weights = None
        self.w_0 = None
        self.classes = None

    def logistic_function(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, Y):
        X = np.array(X)
        Y = np.array(Y)
        self.classes = np.unique(Y)
        n_samples, n_features = X.shape
        n_classes = len(self.classes)

        self.w_0 = np.zeros(n_classes)
        self.weights = np.zeros((n_classes, n_features))


        for idx, cls in enumerate(self.classes):
            y_binary = (Y == cls).astype(int)  
            weights = np.zeros(n_features)
            w_0 = 0

            for i in range(self.iterations):
                linear_model = np.dot(X, weights) + w_0
                y_pred = self.logistic_function(linear_model)

                gradient_weights = np.dot(X.T, (y_pred - y_binary)) / n_samples
                gradient_w_0 = np.sum(y_pred - y_binary) / n_samples

                weights -= self.learning_rate * gradient_weights
                w_0 -= self.learning_rate * gradient_w_0

                if np.linalg.norm(gradient_weights) < self.tolerance and abs(gradient_w_0) < self.tolerance:
                    break

            self.weights[idx] = weights
            self.w_0[idx] = w_0

        print(f"Training completed for {n_classes} classes.")

    def predict(self, X):
        X = np.array(X)
        linear_models = np.dot(X, self.weights.T) + self.w_0
        probabilities = self.logistic_function(linear_models)
        return self.classes[np.argmax(probabilities, axis=1)]
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

X, Y = make_classification(n_samples=1000, n_features=10, random_state=42)                       

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)


LogReg = LogisticRegression()

LogReg.fit(X_train, Y_train)

predictions = LogReg.predict(X_test)

print(f"Classification Report:\n{classification_report(Y_test, predictions)}")
```
